# Importing data and visualize 

```python
concrete_data = pd.read_csv("concrete_data.csv")
concrete_data.head()
```

```python
## Split the data into predictor and targets 
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] 
n_cols = len(predictors.columns)
```

```python
from sklearn.model_selection import train_test_split
```

#Model 

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
def run_task_1():
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.30 )
    model = regression_model()
    model.fit(X_train, y_train, validation_split=0.3, epochs=50, verbose=0)
    mse = model.evaluate(X_test, y_test, verbose=0)
    return mse

mse_list = []
# The task is run 50 times to obtain a list of mse 
for i in range(50):
    mse = run_task_1()
    mse_list.append(mse)
```

# Result of task A

```python
## Result of task A. MSe error has been generated for ll 50 times the mse was generated 
print("Mean of the mean squared error:", np.mean(np.array(mse_list)))
print("Standard deviation of the mean squared error:", np.std(np.array(mse_list)))
```

# Normalized data frame - task B

```python
from sklearn.preprocessing import StandardScaler

normalized_pred = pd.DataFrame(StandardScaler().fit_transform(predictors), columns=predictors.columns)

mean_value = np.mean(target)
std_value = np.std(target)

# Z-score normalization
normalized_target= (target - mean_value) / std_value
```

```python
def run_task_B():
    X_train, X_test, y_train, y_test = train_test_split(normalized_pred, normalized_target, test_size=0.30 )
    model = regression_model()
    model.fit(X_train, y_train, validation_split=0.3, epochs=50, verbose=0)
    mse = model.evaluate(X_test, y_test, verbose=0)
    return mse

mse_list_nornaliz = []
# The task is run 50 times to obtain a list of mse 
for i in range(50):
    mse = run_task_B()
    mse_list_nornaliz.append(mse)
```

```python
## Result of task A. MSe error has been generated for ll 50 times the mse was generated 
print("Mean of the mean squared error:", np.mean(np.array(mse_list_nornaliz)))
print("Standard deviation of the mean squared error:", np.std(np.array(mse_list_nornaliz)))
```
